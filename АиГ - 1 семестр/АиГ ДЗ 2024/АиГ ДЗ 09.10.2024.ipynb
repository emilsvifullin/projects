# Задание 5.1

```python
#1 вариант решения
A = np.array([[ 1,  2,  3],
              [ 0, -1,  1],
              [-2,  1,  0],
              [-1,  2, -3]])

P23 = np.eye(4)
P23[[1, 2]] = P23[[2, 1]]

B = P23 @ A

B[2] = B[2] + (-2) * B[3]
B[2] = -2 * B[2]
B[3] = B[3] + B[1]
B[[0, 3]] = B[[3, 0]]

b32 = B[2, 1]
b23 = B[1, 2]

Matrix(B), b32, b23
```

```python
#2 вариант решения
A = Matrix([[1, 2, 3],[0, -1, 1],[-2, 1, 0],[-1, 2, -3]])

P23 = sp.eye(4).elementary_row_op('n<->m', row1=1, row2=2)
P14 = sp.eye(4).elementary_row_op('n<->m', row1=0, row2=3)
A34 = sp.eye(4).elementary_row_op('n->n+km', row1=2, row2=3, k=-2)
M3 = sp.eye(4).elementary_row_op('n->kn', row=2, k=-2)
A42 = sp.eye(4).elementary_row_op('n->n+km', row1=1, row2=3, k=1)

B = P23 * A34 * M3 * A42 * P14 * A

b32 = B[2, 1]
b23 = B[1, 2]

B, b32, b23
```

# Задание 5.4

```python
A_inv = Matrix([[2, 1], [3, 1]])
C_inv = Matrix([[0, 2], [1, 4]])
b = Matrix([1, 1])

A = A_inv.inv()
C = C_inv.inv()

x1, x2 = symbols('x1 x2')

X = Matrix([x1, x2])

eq = Eq(C.T * A * X, b)

res = solve(eq, [x1, x2])

x1_value = res[x1]
x2_value = res[x2]

print(f"x1 = {x1_value}, x2 = {x2_value}")
```
