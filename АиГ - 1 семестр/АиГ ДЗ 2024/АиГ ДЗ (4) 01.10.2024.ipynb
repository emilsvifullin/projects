# Задание 4.6

```python
b11,b12,b21,b22 = symbols('b11 b12 b21 b22')
A = Matrix([[1,2],[3,4]])
B = Matrix([[b11,b12],[b21,b22]])
AB = A*B
BA = B*A
eqs = [Eq(AB[i, j], BA[i,j]) for i in range(2) for j in range(2)]
solution = solve(eqs, [b11,b12,b21,b22])
solution
```
